```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
neighborhood = air_bnb.groupby('neighbourhood_group', as_index=False)
neighborhood.count()[['neighbourhood_group', 'neighbourhood']].sort_values(['neighbourhood'], ascending=False)

# Answer: Manhattan = 21661 neighborhood groups
```

```python
# Are private rooms the most popular in manhattan?
neighborhood = air_bnb.groupby(['neighbourhood_group', 'room_type'], as_index=True)
neighborhood[['room_type']].count()

# Answer: No, Entire home/apt are the most popular in Manhattan.
```

```python
# Which hosts are the busiest and based on their reviews?
host = air_bnb.groupby('host_name', as_index=True)
host[['number_of_reviews']].count().sort_values(['number_of_reviews'], ascending=False)

# Answer: Michael is the most busiest host. He has a total of 417 reviews.
```

```python
#Which neighorhood group has the highest average price?
neighborhood = air_bnb.groupby('neighbourhood_group', as_index=True)
neighborhood[['price']].mean().sort_values(['price'], ascending=False)

# Answer : Manhattan has the highest avg price at $196.88
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.sort_values(['price'], ascending=False)[['neighbourhood_group','price']]

# Answer: Queens, Brooklyn, and Manhattan are tied for highest total price at $10,000.
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.sort_values(['price'], ascending=False)[['host_name', 'price']].head()

# Answer: Kathrine, Erin, Jelena, Matt, and Amy have the highest total price.
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb[(air_bnb['availability_365']==0) & (air_bnb['number_of_reviews'] >= 100)][['host_name', 'availability_365']].reset_index()

# Answer: 161 hosts with a review count of 100 or more have zero availability
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.sort_values(['price'], ascending=False)[['host_name', 'price', 'neighbourhood']]

# Answer: Kathrine, Erin, Jelena all have a total of prices of 10000, located in Astoria, Greenpoint, and Upper West Side.
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')][['host_name','last_review']].sort_values('last_review', ascending=False)

# Answer: Danielle's last review was on 2019-07-08.
```

```python
air_bnb.groupby('room_type').sum('number_of_reviews')[['number_of_reviews']]

# Answer: Entire home/apt have the highest review numbers.
```
