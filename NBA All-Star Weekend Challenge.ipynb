```python
import pandas as pd
import numpy as np

nbaStats = pd.read_csv('./NBA All Stars 2000-2016 - Sheet1.csv')
nbaStats
```

# What is the average weight of all players who played during this time?

```python
avgWeight = nbaStats.drop_duplicates(['Player'])
avgWeight['WT'].mean()

# Answer: 228 lbs
```

# What team is represented the most during this time?

```python
mostRep = nbaStats.groupby('Team').count().sort_values('Year', ascending=False)
mostRep

# Answer: Miami Heat
```

# What draft class is represeted the most during this time?

```python
mostRepClass = nbaStats.groupby('NBA Draft Status').count().sort_values('Player', ascending=False)
mostRepClass

# Answer: Class of 1996 Rnd 1, Pick 13
```

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
nbaStats.groupby(['Nationality']).count().sort_values('Player', ascending=False)

# Answer: United States is much more prevalent in the All-Star games.
```

# How often are the Western All-Stars voted in by fans?

```python
nbaStats[nbaStats['Selection Type'] == 'Western All-Star Fan Vote Selection'].groupby('Year').count()

# Answer: Every year is 5.
```

# How many times does Steph Curry make the All-Star game during this time period?

```python
nbaStats[nbaStats['Player']=='Stephen Curry']

# Answer: 4 Times
```

# How many Shooting Guards have made the All-Star game during this time?

```python
nbaStats.drop_duplicates(['Player']).groupby('Pos').count()[['Player']]

# Answer: There have been 9 different shooting guards during this time.
```

# How many Chicago Bulls players have made the All-Star Game during this time period?

```python
noDuplicates = nbaStats.drop_duplicates(['Player'])
noDuplicates[noDuplicates['Team']=='Chicago Bulls'].count()[['Player']]

# Answer: 5 different Chicago Bulls players have made the All-Star game durign this period.
```
